```python
#accuracy, precision, and recall
from sklearn.metrics import accuracy_score, recall_score, precision_score

df = pd.read_csv("Comments_Sentiments.csv")
predictions = df["predictions"] #positive, negative, neutral
actual_labels = df["sentiments"] #positive, negative, neutral
correct_incorrect_prediction = df["prediction-correct(0/1)"] #positive = 1, negative = 0
correct_incorrect_actual = df["prediction-correct-actual(0/1)"] #positive = 1, negative = 0

accuracy = accuracy_score(predictions, actual_labels)
recall = recall_score(correct_incorrect_prediction, correct_incorrect_actual)
precision = precision_score(correct_incorrect_prediction, correct_incorrect_actual)

print("Accuracy: ", accuracy)
print("Recall: ", recall)
print("Precision: ", precision)
```
